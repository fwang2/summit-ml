# Matplotlib




````python
import numpy as np 
import matplotlib as mpl 
import matplotlib.pyplot as plt 
%matplotlib inline 
x = np.linspace(0, 10, 100)
````

## Draw a simple plot and save


````python

fig = plt.figure() 
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
plt.savefig('myfig.pdf')
````

## Object oriented interface

````python
# ax will be an array of two axes object
fig, ax = plt.subplots(2)
x = np.linspace(0, 10, 100)
ax[0].plot(x, np.sin(x), '-')
ax[1].plot(x, np.cos(x), '--')
````

## change plotting style

## Adjust line color and styles

use `plot( ..., color='red, linestyle='solid')`. The following example shows shortened version of that.


````python

plt.style.use('seaborn-whitegrid')
fig = plt.figure()
ax = plt.axes()
ax.plot(x, x+0, '-g') # solid green
ax.plot(x, x+1, '--c') # dashed cyan
ax.plot(x, x+2, ':r') # dotted red
````

## Adjust Axes limits

This is done by
```
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
```

You can "tighten" the bound of axis as it usually leave some white space around:

```
plt.axis('tight')
plt.axis('equal')
```
the second command will adjust aspect ratio to be "equal" for x and y axis.


## Labeling plot

you can use `plt.plot( ..., label="sin(x)") to label a plot. Or define them separately.


```
plt.title("My title")
plt.xlabel("x")
plt.ylabel('y')
plt.legend()
```

The last command above is to show the legend.




## Scatter plot

You can supply a 3rd parameter to `plot` command to make it a scatter plot instead of line plot. For example: `plt.plot(x, y, 'o', color='black')`. A second and more powerful command is `plt.scatter(x,y, marker=...)` as show below.

The following example shows various marker you can use (From VanderPls Python Data book)



````python
#plt.figure(figsize=(10,6))
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.scatter(rng.rand(5), rng.rand(5), marker=marker,
        label=f"marker={marker}")
plt.legend(numpoints=1)
plt.xlim(0, 1.8)
````

### Customize size, color and transparency in scatter plot


````python
rng = np.random.RandomState(0)
x = rng.randn(100) # normal distribution
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000*colors # the example from book seems off: sizes = 1000*rng.rand(1000)
plt.scatter(x,y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar()
````

### using iris dataset, viz 4 dimension data

````python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

# iris.feature_names
#
# feature[0]: septal length
# feature[1]: septal width
# feature[2]: petal length
# feature[3]: petal width
# target: species of the flower

# in the following code, we map:
# feature[0] to x
# feature[1] to y
# feature[3] to size
# feature[3] to color

plt.scatter(features[0], features[1], alpha=0.2,
    s=100*features[3], c=iris.target, cmap='viridis')

plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.colorbar()
````

## Errors


### Basic errorbar

the following code make errorbars lighter than points themselves.


````python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy*np.random.randn(50)
plt.errorbar(x,y, yerr=dy, fmt='o', color='black',
    ecolor='lightgray', elinewidth=3, capsize=4)
````
